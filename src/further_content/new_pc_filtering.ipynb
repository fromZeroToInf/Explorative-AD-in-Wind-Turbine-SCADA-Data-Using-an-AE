### Changing the existing pc filtered datapoints
(instead of re-run the whole preprocessing step)

```python
from pathlib import Path
import pandas as pd
from helperfunctions import intern_constants as ic
import os
import shutil
from pathlib import Path
```

```python

SRC = Path(ic.PATH_PC_FILTERING)
os.makedirs(SRC / "backup", exist_ok=True)

for fp in SRC.glob("*csv"):
    shutil.copy2(fp, SRC/"backup"/ fp.name)
```

```python
wind_col = "Wind speed (m/s)"
power_col = "Power (kW)"

min_wind = 5.0
min_power = 50
```

```python
for fp in Path(SRC).glob("*.csv"):
    df = pd.read_csv(fp, parse_dates=[ic.TS_COL])
    df = df[(df[ic.TS_COL] >= ic.TRAIN_START) & (df[ic.TS_COL] <= ic.TRAIN_END)]
    df[wind_col] = df[wind_col].astype(float)
    df[power_col] = df[power_col].astype(float)
    df =df[(df[wind_col] > min_wind) & (df[power_col] > min_power)]
    df.to_csv(SRC / fp.name, index=False)
```
