### Preprocessing step 1

Aggregate wind turbine data per year

Obsolete folders after execution will be deleted

```python
from pathlib import Path
import os
from glob import glob
from tqdm import tqdm
import shutil
from helperfunctions import intern_constants as ic
```

Put all wind turbine scada data (csv files) into the Penmashiel folder

```python
path = Path(ic.PATH_PROJECT_ROOT / "src" / "Penmanshiel")
paths = [fn for fn in os.listdir(path)]
print(paths)
folder_dir = ""
for index in tqdm(range(len(paths))):
    if index % 2 ==0:
        folder_dir = os.path.join(path, paths[index])
    else:
        folder_to_cpy = os.path.join(path, paths[index])
        csv_files = glob(os.path.join(folder_to_cpy, "*.csv"))
        for file in tqdm(csv_files):
            filename = os.path.basename(file)
            Path(os.path.join(folder_dir, filename)).write_bytes(Path(file).read_bytes())
        # Delete this folder, this folder is obsolete.
        if Path(folder_to_cpy).exists() and Path(folder_to_cpy).is_dir():
            shutil.rmtree(folder_to_cpy)
        else:
            print(f"Error in deleting obsolete folder:{folder_to_cpy}")
```
